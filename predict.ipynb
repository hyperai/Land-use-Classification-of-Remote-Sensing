```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

plt.figure(figsize=(20, 20))
for idx in range(3):
    img = Image.open(f"/openbayes/home/semantic_pytorch/out/result/pic3/{idx}.png")
    label = Image.open(f"/openbayes/home/semantic_pytorch/out/result/label/{idx}.png")
    pred = Image.open(f"/openbayes/home/semantic_pytorch/out/result/label/{idx}.png")
    mask = Image.open(f"/openbayes/home/semantic_pytorch/out/result/mask/{idx}.png")

    plt.subplot(4, 4, idx*4+1)
    plt.title('pic')
    plt.imshow(img)
    plt.axis('off')

    plt.subplot(4, 4, idx*4+2)
    plt.title('label')
    plt.imshow(label)
    plt.axis('off')

    plt.subplot(4, 4, idx*4+3)
    plt.title('predict')
    plt.imshow(pred)
    plt.axis('off')

    plt.subplot(4, 4, idx*4+4)
    plt.title('mask')
    plt.imshow(mask)
    plt.axis('off')
plt.show()
```
